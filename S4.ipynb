##1. Finnið meðalfermetraverð á höfuðborgarsvæðinu fyrir fjölbýli á bilinu 60-120m2

```python
#Lesum inn skránna
df = pd.read_csv("kaupskra.csv",encoding='iso-8859-1',sep=';')

#skilyrðin
tafla=df.loc[(df['TEGUND']=="Fjölbýli") &
 ((df['SVFN']==0.0) | (df['SVFN']==1100.0)|(df['SVFN']==1000.0)|(df['SVFN']==1300.0)|
  (df['SVFN']==1400.0)|(df['SVFN']==1604.0)) & (df['EINFLM']<=120) & (df['EINFLM']>=60)]

#sýnum töfluna sem hefur línurnar
#sem uppfylla skilyrðin
display(tafla)

#reiknum fasteignaverð fyrir hverja
#fasteign í töflunni
fverd=tafla.KAUPVERD/tafla.EINFLM

#finnum meðaltalið af fasteignaverði og prentum
print(f"Þannig meðalfermetraverðið er {fverd.mean()}")
```
